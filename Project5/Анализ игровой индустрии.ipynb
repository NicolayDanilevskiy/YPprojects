Расшифровок кодов рейтинга в предоставленных данных нет. Воспользуемся открытыми данными(wikipedia: https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board) и создадим соответствующую переменную.
- `e` - (Everyone) — Для всех. Содержание вполне подходит для возрастной категории от 6 лет.
- `m` - (Mature) — Для взрослых. Материалы игры не подходят для подростков младше 17 лет.
- `t` - (Teen) — Подросткам. Игра подходит для лиц от 13 лет.
- `e10+` - (Everyone 10 and older) — Для всех от 10 лет и старше.
- `k-a` - (Kids to Adults) - Для всех. Устаревшая категория, можно заменить на `e`.
- `ao` - (Adults Only 18+) — Только для взрослых. Содержание игры только для взрослых старше 18 лет.
- `ec` - (Early childhood) — Для детей младшего возраста. Игра подходит для детей от 3 лет.
- `rp` - (Rating Pending) — Рейтинг ожидается/ Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Можно заменить на пропуск, по факту рейтинг продукту не присвоен.

```python
# Произведем замену данных в столбце.
data['rating'] = data['rating'].replace('rp', np.NaN)
data['rating'] = data['rating'].replace('k-a', 'e')
```

**Изучим жизненный цикл игровых платформ.**

```python
# Определим годы выпуска первой и последней игры для каждой платформы.
platform_list = data['platform'].unique().tolist()
first_year = []
last_year = []
lifetime_data = []
for platform in platform_list:    
    year_min = data.loc[data['platform'] == platform,'year_of_release'].min()
    year_max = data.loc[data['platform'] == platform,'year_of_release'].max()
    lifetime = len(data.loc[data['platform'] == platform,'year_of_release'].unique())
    first_year.append(year_min)
    last_year.append(year_max)
    lifetime_data.append(lifetime)
```

## Проверка гипотез.

- Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One(`xone`) и PC(`pc`) одинаковые.
- Гипотеза 2: Средние пользовательские рейтинги жанров `action` и `sports` разные.

Так как выборки у нас независимые, воспользуемся функцией ttest_ind(t-критерий Стьюдента).

### Гипотеза 1.

- H0 - Cредние пользовательские рейтинги платформ `xone` и `pc` одинаковые.
- H1 - Cредние пользовательские рейтинги платформ `xone` и `pc` разные.

<div class="alert alert-block alert-success">
 
### Комментарий ревьюера
    
#### Успех
    
Гипотезы сформированы и метод определен верно. Можем приступать дальше 
</div>

```python
xone_rating = actual_data.loc[(actual_data['platform'] == 'xone') & (actual_data['user_score'].notna())]['user_score']
pc_rating = actual_data.loc[(actual_data['platform'] == 'pc') & (actual_data['user_score'].notna())]['user_score']
```

```python
xone_rating
```

```python
pc_rating
```

```python
# Проверим гипотезу с помощью ttest_ind.
alpha = .05 # критический уровень статистической значимости, если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(xone_rating, pc_rating, equal_var = False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное p-значение не позволяет отвергнуть нулевую гипотезу и говорит о том, что "Cредние пользовательские рейтинги платформ `xone` и `pc` одинаковые". Вероятность получить подтверждение нулевой гипотезы 55%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между рейтингами платформ `xone` и `pc`.
