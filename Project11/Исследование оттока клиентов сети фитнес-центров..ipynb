```python
# –ó–∞–¥–∞–¥–∏–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

rf_model = RandomForestClassifier(n_estimators = 100,random_state = 0)

# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

rf_model.fit(X_train_st, y_train)

# –í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —É–∂–µ –æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑—ã

rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

### –û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision, recall, f1 –∏ roc_auc –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏–º –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –í—ã—è—Å–Ω–∏–º –∫–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?

- Accuracy: –¥–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤
- Precision: —Ç–æ—á–Ω–æ—Å—Ç—å (–∞–Ω–≥–ª. precision)
- Recall: –ø–æ–ª–Ω–æ—Ç–∞
- F1_score: –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É Precision and Recall
- ROC_AUC: –¥–æ–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞

```python
# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã–≤–æ–¥–∞ –º–µ—Ç—Ä–∏–∫.

def print_all_metrics(y_true, y_pred, y_proba, title):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º –≤–µ–∫—Ç–æ—Ä –ø—Ä–æ–≥–Ω–æ–∑–æ–≤ —Å —É—á—ë—Ç–æ–º –ø–æ—Ä–æ–≥–∞, —Ä–∞–≤–Ω–æ–≥–æ 0.4. C—Ä–∞–≤–Ω–∏–º —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ –ø—Ä–∏ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–º —Ä–∞—Å—á—ë—Ç–µ –∫–ª–∞—Å—Å–æ–≤.

```python
# –í—ã–±–µ—Ä–µ–º –≤—Ä—É—á–Ω—É—é —É—Ä–æ–≤–µ–Ω—å –æ—Ç—Å–µ—á–µ–Ω–∏—è.

threshold = 0.4

# –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º predict.

lr_custom_predictions = [0 if i<threshold else 1 for i in lr_probabilities]
rf_custom_predictions = [0 if i<threshold else 1 for i in rf_probabilities]
```

```python
# –í—ã–≤–µ–¥–µ–º –≤—Å–µ –º–µ—Ç—Ä–∏–∫–∏.

print_all_metrics(y_test, lr_predictions, lr_probabilities, '–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:')
print_all_metrics(y_test, rf_predictions, rf_probabilities, '–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:')
print_all_metrics(y_test, lr_custom_predictions, lr_probabilities,
                  '–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –∫–∞—Å—Ç–æ–º–Ω—ã–º –ø–æ—Ä–æ–≥–æ–º:')
print_all_metrics(y_test, rf_custom_predictions, rf_probabilities,
                  '–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ —Å –∫–∞—Å—Ç–æ–º–Ω—ã–º –ø–æ—Ä–æ–≥–æ–º:')
```

```python
# –°—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –∏ —Ä–∏—Å—É–µ–º –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—É.

linked = linkage(x_sc, method = 'ward')
plt.figure(figsize=(20, 15))  
dendrogram(linked, orientation='top')
plt.title('–ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
plt.show()
```

### –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ K-Means –∏ —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã –∫–ª–∏–µ–Ω—Ç–æ–≤.  –î–æ–≥–æ–≤–æ—Ä–∏–º—Å—è –∑–∞ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –ø—Ä–∏–Ω—è—Ç—å n=5.

```python
# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É K-Means.

km = KMeans(n_clusters = 5, random_state=0)
labels = km.fit_predict(x_sc)
data_for_model['cluster'] = labels
```
